```python
import numpy as np
import cv2
```

```python
import insightface
```

```python
from insightface.app import FaceAnalysis
```

### Buffalo_l model

```python
app_l = FaceAnalysis(name = 'buffalo_l',
                     root = 'insightface_model',
                     providers=['CPUExecutionProvider']
                    )
app_l.prepare(ctx_id=0, det_size=(640,640))
```

### Buffalo_sc model

```python
app_sc = FaceAnalysis(name = 'buffalo_sc',
                     root = 'insightface_model',
                     providers=['CPUExecutionProvider']
                    )
app_sc.prepare(ctx_id=0, det_size=(640,640))
```

### Load images 

```python
img = cv2.imread('test_image_2.jpg')

cv2.imshow('image', img)
cv2.waitKey(0)# as soon as if i press esc key or x button, then my window will get closed
cv2.destroyAllWindows()
```

### Apply image to insightface buffalo_l model and get results

```python
results_l = app_l.get(img)
```

```python
print(results_l)
```

```python
results_l[0]['det_score']
```

```python
img_copy = img.copy()
gender_encode = ['Female', 'Male']
for res in results_l:
    x1, y1, x2, y2 = res['bbox'].astype(int)

    # draw a bounding box 
    cv2.rectangle(img_copy, (x1,y1), (x2,y2), (0,255,0), 2)

    #draw a circle at kps
    kps =res['kps'].astype(int)
    for k1, k2 in kps:
        cv2.circle(img_copy,(k1,k2),2,(0,255,255),-1)

    #detection score
    score = "score : {}%".format(round(res['det_score']*100, 2))
    cv2.putText(img_copy,score,(x1,y1-1),cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,0,255))

    #age and gender
    gender = gender_encode[res['gender']]
    age = res['age']
    age_gender = f"{gender}::{age}"

    cv2.putText(img_copy,age_gender,(x1,y2+12),cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,0,255))

cv2.imshow('image', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### For buffalo_sc model

```python
app_sc = FaceAnalysis(name = 'buffalo_sc',
                     root = 'insightface_model',
                     providers=['CPUExecutionProvider']
                    )
app_sc.prepare(ctx_id=0, det_size=(640,640))
```

```python
result_sc = app_sc.get(img)
```

```python
len(result_sc)
```

```python
result_sc[0].keys()
```

```python
img_copy = img.copy()
for res in result_sc:
    #bounding box
    x1, y1, x2, y2 = res['bbox'].astype(int)
    cv2.rectangle(img_copy,(x1,y1),(x2,y2),(0,255,0),2)

    #kps
    kps = res['kps'].astype(int)
    for kp in kps:
        cv2.circle(img_copy,kp,2,(0,255,255), -1)

    #detection score
    score = "score : {}%".format(int(res['det_score']*100))
    cv2.putText(img_copy,score,(x1,y1),cv2.FONT_HERSHEY_DUPLEX,0.7,(255,0,0),2)

cv2.imshow('image', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
